```
import pandas as pd
import numpy as np
```

```
df = pd.read_csv('/content/train.csv', usecols=['Survived', 'Cabin'])
df.head()
```

**Targeted Ordinal Encoding**

```
df['Cabin'].fillna('Missing', inplace=True)
```

```
df.head()
```

```
df['Cabin'] = df['Cabin'].astype('str').str[0]
```

```
df['Cabin'].unique()
```

```
df.groupby(['Cabin'])['Survived'].mean()
```

```
order = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
order
```

```
dictionary = {'T':1, 'M':2, 'A':3, 'G':4, 'C':5, 'F':6, 'B':7, 'E':8, 'D':9}
```

```
df['Targeted_Ordinal_Encoding'] = df['Cabin'].map(dictionary)
```

```
df.head()
```

Mean Encoding

```
mean_ecoding=df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```
df['Mean_Encoding'] = df['Cabin'].map(mean_ecoding)
```
