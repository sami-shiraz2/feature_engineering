```
import pandas as pd

df = pd.read_csv('/content/train.csv', usecols=['Survived', 'Cabin'])
```

```
df.head()
```

```
df['Cabin'].fillna('Missing', inplace=True)
```

```
df.head()
```

```
df['Cabin'] = df['Cabin'].astype('str').str[0]
```

```
df.head()
```

```
m=df.groupby('Cabin')['Survived'].mean()
```

```
data = pd.DataFrame(m)
```

```
data
```

```
data['Died'] = 1-data['Survived']
```

```
data.head()
```

```
data['Probability_Ratio'] = data['Survived']/data['Died']
```

```
data.head()
```

```
dictt = data['Probability_Ratio'].to_dict()
```

```
dictt
```

```
df['Cabin_Encoded'] = df['Cabin'].map(dictt)
```

```
df.head()
```
